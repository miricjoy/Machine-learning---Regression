```python
#pandas library contains functions for database handling- opening csv(comma seperated version) files and data handling
import pandas as pd
```

```python
#read the data file from the specified location using read_csv and save the dataframe in 'dataset' variable
dataset=pd.read_csv("/Users/stalin/Desktop/hope AI/Machine learning /linear regression/Salary_Data.csv")
```

```python
#dataset is a dataframe 
type(dataset)
```

```python
#viewing the dataset - it gives salary details for corresponding years of experience
dataset
```

```python
# assigning yearsExperience column as independent/input variable
x=dataset[["YearsExperience"]]
```

```python
#assigning salary column as output/dependent variable
y=dataset[["Salary"]]
```

```python
# scikit learn - sklearn library contains methods for machine learning algorithms
#test_train split - split the dataset into training set with 70% data and test set with 30% data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
print(x_train)
type(x_train)
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#importing Linear Regression method -used for data which has linear behaviour
from sklearn.linear_model import LinearRegression
```

```python
#create regressor object for Linear regression class
regressor=LinearRegression()
#fit method is used to train the given data and find the slope and intercept of the trend line and store it in regressor object
regressor.fit(x_train,y_train)
```

```python
#coef_ stores the weight/slope of the trend line
weight=regressor.coef_
```

```python
#intercept_ stores the y_intercept / origin value of the trend line
bias=regressor.intercept_
```

```python
print(f"The weight of the model is {weight}")
print(f"The bias of the model is {bias}")
```

```python
#using the test set to predict the salary output for given years of experience input
y_pred=regressor.predict(x_test)
print(y_pred)
type(y_pred)
```

```python
#using r^2 metric to find accuracy of the model ,by comparng predicted values with test data values
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
#save the trained model to use for deployment using pickle library
import pickle
filename="finalised_model.sav"
#dump method is used to write the trained model in finalised_model file
pickle.dump(regressor,open(filename,"wb"))
```

```python
#for deployment, read the saved model using pickle.load method
loaded_model=pickle.load(open("finalised_model.sav","rb"))
# use the loaded model to predict salary for 'yearsExperience' input
# for x=0, yearsExperience=0, we get the intercept value as output
result=loaded_model.predict([[0]])
result
```

```python
#get user input for YearsExperience
get_input=int(input("Enter no. of years of Experience"))
```

```python
#predict the salary for given user input
predicted_salary=loaded_model.predict([[get_input]])
print(f"The predicted salary is {predicted_salary}")
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
dataset
```

```python
dataset.columns
```

```python
x=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
x
```

```python
y=dataset['Profit']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random')

regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
rscore=r2_score(y_pred,y_test)
rscore
```

```python
import pickle
pickle.dump(regressor,open("finalised_dt_model.sav","wb"))
```
